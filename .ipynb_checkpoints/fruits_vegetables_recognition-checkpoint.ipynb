```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense,Conv2D,Flatten,MaxPooling2D,Input,Dropout
from tensorflow.keras.layers import LeakyReLU
```

## Preprocessing the dataset

```python

# Preprocessing the Training set
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('C:\\Users\\ASUS\\Downloads\\archive (5)\\train',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')

# Preprocessing the valid  set
valid_datagen = ImageDataGenerator(rescale = 1./255)
valid_set = valid_datagen.flow_from_directory('C:\\Users\\ASUS\\Downloads\\archive (5)\\validation',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
# Preprocessing the test set
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('C:\\Users\\ASUS\\Downloads\\archive (5)\\test',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

```python
training_set.image_shape
```

```python
test_set.image_shape
```

```python
valid_set.image_shape
```

## Applying the model

```python
model = tf.keras.Sequential([
    Input(shape=[64,64,3]),
    Conv2D(filters=32,kernel_size=3,activation=LeakyReLU(alpha=0.05)),
    MaxPooling2D(pool_size=3, strides=2),
    Flatten(),
    Dense(200,activation=LeakyReLU(alpha=0.05)),
    Dense(100,activation=LeakyReLU(alpha=0.05)),
    Dense(36,activation="softmax")
])
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss="sparse_categorical_crossentropy",optimizer=opt,metrics=["accuracy"])
# Fitting the model
r= model.fit(training_set,epochs=20,validation_data=valid_set)
```

```python
model.evaluate(valid_set)
```

```python
# evaluating with test set
model.evaluate(test_set)
```

```python
import matplotlib.pyplot as plt
# Plot training & validation accuracy values
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(r.history['accuracy'])
plt.plot(r.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
```

```python
plt.plot(r.history['loss'])
plt.plot(r.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
```
